## Edge index 생성

```python
# 장르로 edge를 연결
unique_genres_combinations = final_metadata_with_overview['genres'].unique()

unique_genres = set() # 고유한 장르를 저장할 집합

for combination in unique_genres_combinations:
    genres = combination.split(',') # 장르 조합을 분리
    unique_genres.update(genres) # 집합에 추가하여 고유한 장르만을 보존
unique_genres = unique_genres - {'\\N'}
sorted(unique_genres)
```

```python
# 각 영화의 인덱스를 키로, 해당 영화의 장르를 값으로 가지는 딕셔너리 생성
movie_genres = {}
for index, row in final_metadata_with_overview.iterrows():
    genres = row.iloc[5].split(',')  # ','로 장르 분리
    movie_genres[index] = set(genres)  # 장르 집합 생성하여 딕셔너리에 저장
# movie_genres
# 각 장르별 영화 개수 plot
import matplotlib.pyplot as plt

genre_count = {}
for genres in movie_genres.values():
    for genre in genres:
        if genre in genre_count:
            genre_count[genre] += 1
        else:
            genre_count[genre] = 1
genre_count = dict(sorted(genre_count.items(), key=lambda x: x[1], reverse=True))
plt.figure(figsize=(20, 10))
plt.bar(genre_count.keys(), genre_count.values())
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np

# 각 영화의 인덱스를 키로, 해당 영화의 장르를 값으로 가지는 딕셔너리 생성
movie_genres = {}
for index, row in final_metadata_with_overview.iterrows():
    genres = row.iloc[5].split(',')  # ','로 장르 분리
    movie_genres[index] = set(genres)  # 장르 집합 생성하여 딕셔너리에 저장

# edge_index 생성
edge_index = [[], []]

# 영화 쌍을 비교하여 공통 장르가 있는 경우 edge_index에 추가
for movie1 in movie_genres.keys():
    for movie2 in movie_genres.keys():
        if movie1 < movie2:  # 중복 비교 방지
            # 공통 장르 확인
            if not movie_genres[movie1].isdisjoint(movie_genres[movie2]):
                # 두 영화 간 공통 장르가 있으면 edge_index에 추가
                edge_index[0].append(movie1)
                edge_index[1].append(movie2)
                # 양방향 연결을 위해 반대 방향도 추가
                edge_index[0].append(movie2)
                edge_index[1].append(movie1)
```

```python
edge_index = np.array(edge_index)
edge_index.shape
# edge_index의 edge_index[0]이 1인 것을 추출
# edge_index[:, edge_index[0] == 1].shape
```

```python
# genres_edge_index 저장
np.save('genres_edge_index.npy', edge_index)
```

```python
# 감독으로 edge를 연결

import numpy as np

unique_directors_combinations = final_metadata_with_overview['directors'].unique()

unique_directors = set()

for combination in unique_directors_combinations:
    directors = combination.split(',')
    unique_directors.update(directors)

# 각 영화의 인덱스를 키로, 해당 영화의 감독을 값으로 가지는 딕셔너리 생성
movie_directors = {}
for index, row in final_metadata_with_overview.iterrows():
    directors = row.iloc[8].split(',')  # ','로 감독 분리
    movie_directors[index] = set(directors)  # 감독 집합 생성하여 딕셔너리에 저장

# edge_index 생성
directors_edge_index = [[], []]

# 영화 쌍을 비교하여 공통 감독이 있는 경우 edge_index에 추가
for movie1 in movie_directors.keys():
    for movie2 in movie_directors.keys():
        if movie1 < movie2:  # 중복 비교 방지
            # 공통 감독 확인
            if not movie_directors[movie1].isdisjoint(movie_directors[movie2]):
                # 두 영화 간 공통 감독이 있으면 edge_index에 추가
                directors_edge_index[0].append(movie1)
                directors_edge_index[1].append(movie2)
                # 양방향 연결을 위해 반대 방향도 추가
                directors_edge_index[0].append(movie2)
                directors_edge_index[1].append(movie1)

directors_edge_index = np.array(directors_edge_index)
directors_edge_index
```
