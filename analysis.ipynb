감독 기준 혹은 배우 기준으로 그래프에 넣을 때 각각의 feature는 어떻게 처리할 것인가

```python
import pandas as pd

custom_metadata = pd.read_csv("custom_metadata_ver1.csv")
custom_metadata
```

```python
actor_string_for_each_movie = custom_metadata.groupby('tconst')['nconst'].apply(lambda x: ','.join(x)).reset_index()
actor_string_for_each_movie
```

```python
# final_metadata에 actor_string_for_each_movie를 merge
final_metadata = custom_metadata.merge(actor_string_for_each_movie, left_on='tconst', right_on='tconst')
final_metadata
```

```python
# final_metadata에서 tconst가 겹치는 데이터 중 첫번째 데이터만 남기고 나머지는 drop
final_metadata = final_metadata.drop_duplicates(['tconst'], keep='first')
final_metadata
```

```python
# nconst_x column drop
final_metadata = final_metadata.drop(columns=['nconst_x', 'category', 'characters'])
final_metadata
```

```python
# final_metadata의 nconst_y column 이름을 actors로 변경
final_metadata.rename(columns={'nconst_y': 'actors'}, inplace=True)
final_metadata
```

```python
final_metadata.to_csv("final_metadata_ver1.csv", index=False)
```
